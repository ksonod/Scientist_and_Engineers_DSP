# Chapter 5: Linear Systems

In this notebook, several decomposition methods are described.

```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import gridspec
```

## Common Decompositions

```python
num_points=50
x=np.linspace(0,10,num_points)

ys1=np.sin(x)
ys2=np.sin(1.5*x)
ys3=np.sin(2*x)
yc1=np.cos(x)
yc2=np.cos(1.5*x)
yc3=np.cos(2*x)


y=ys1+ys2+ys3+yc1+yc2+yc3

fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(x,y)
plt.show()
```

### Impulse Decomposition

```python
y_impulse=np.zeros([num_points,num_points])

for i in range(y.shape[0]):
    y_impulse[i,i]=y[i] #making impulse signals

fig=plt.figure(figsize=(14,9))
ax=fig.add_subplot(511)
ax.plot(x,y, marker='o', color='black') # original signal

ax=fig.add_subplot(512)
ax.plot(x,y_impulse[0], marker='o', color='blue') # first point

ax=fig.add_subplot(513)
ax.plot(x,y_impulse[1], marker='o', color='blue') # second point

ax=fig.add_subplot(514)
ax.plot(x,y_impulse[2], marker='o', color='blue') # third point

ax=fig.add_subplot(515)
ax.plot(x,y_impulse[3], marker='o', color='blue') # fourth point


plt.show()
```

### Even/Odd Decomposition

y<sub>even</sub>[n] = (y[n] + y[N-n]) / 2  
y<sub>odd</sub>[n] = (y[n] - y[N-n]) / 2

```python
y_even=np.zeros_like(y)
y_odd=np.zeros_like(y)

for i in range(num_points):
    y_even[i]= 0.5*(y[i]+y[num_points-i-1])
    y_odd[i]= 0.5*(y[i]-y[num_points-i-1])
    
fig=plt.figure(figsize=(14,5))
ax=fig.add_subplot(111)
ax.plot(x,y, color='black')
ax.plot(x,y_even, color='red')
ax.plot(x,y_odd, color='blue')
ax.legend(['origianl signal','even symmetry', 'odd symmetry'], frameon= False)
plt.show()
```

### Interlaced Decomposition

```python
y_even=np.zeros_like(y)
y_odd=np.zeros_like(y)

for i in range(num_points):
    if i%2==0: # even number
        y_even[i]= y[i]
    else: # odd number
        y_odd[i]= y[i]
    
fig=plt.figure(figsize=(14,5))
ax=fig.add_subplot(111)
ax.plot(x,y, color='black')
ax.plot(x,y_even, color='red')
ax.plot(x,y_odd, color='blue')
ax.legend(['origianl signal','even samples', 'odd samples'], frameon= False)
plt.show()
```

### Fourier Decomposition

```python
fig = plt.figure(figsize = (14,14))

ax1=fig.add_subplot(411)
ax1.plot(x,y, color='black')
ax1.text(8,3,'Original signal')

ax2=fig.add_subplot(423)
ax2.plot(x,ys1, color='black')
ax2.set_title('sine waves')
ax2.text(0,-0.5,'sine 1')

ax3=fig.add_subplot(424)
ax3.plot(x,yc1, color='black')
ax3.set_title('cosine waves')
ax3.text(0,-0.5,'cosine 1')

ax4=fig.add_subplot(425)
ax4.plot(x,ys2, color='black')
ax4.text(0,-0.5,'sine 2')

ax5=fig.add_subplot(426)
ax5.plot(x,yc2, color='black')
ax5.text(0,-0.5,'cosine 2')

ax6=fig.add_subplot(427)
ax6.plot(x,ys3, color='black')
ax6.text(0,-0.5,'sine 3')

ax7=fig.add_subplot(428)
ax7.plot(x,yc3, color='black')
ax7.text(0,-0.5,'cosine 3')

plt.show()
```
