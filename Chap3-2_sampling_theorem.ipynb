```python
import numpy as np
import matplotlib.pyplot as plt
```

# Chapter 3 (part2) Sampling theorem

Input analog signal is shown below. This is sinusoidal curve.

```python
tim_max=3000

x=np.linspace(0,tim_max,1001)
y=np.sin(x/180*np.pi)
analog_freq=(1/360) # analog frequency

fig=plt.figure(figsize=(14,4))
ax=fig.add_subplot(111)

ax.plot(x,y)

plt.show()
```

Data with different sampling rates are shown below.

```python
fig=plt.figure(figsize=(18,8))
ax=fig.add_subplot(221)

ax.plot(x,y, color='black')
num_samples=90
sampling_x = np.linspace(0,num_samples*((y.shape[0]-1)//num_samples), num_samples).astype(np.int16) 
ax.plot(x[sampling_x] , y[sampling_x], marker='o', linewidth=0, color='red')
sample_freq=1/(x[sampling_x][1]-x[sampling_x][0])
ax.set_xlim([0,2500])
ax.set_title("analog_freq/sample_freq=%0.3f"%(analog_freq/sample_freq))

ax=fig.add_subplot(222)

ax.plot(x,y, color='black')
num_samples=28
sampling_x = np.linspace(0,num_samples*((y.shape[0]-1)//num_samples), num_samples).astype(np.int16) 
ax.plot(x[sampling_x] , y[sampling_x], marker='o', linewidth=0, color='red')
sample_freq=1/(x[sampling_x][1]-x[sampling_x][0])
ax.set_xlim([0,2500])
ax.set_title("analog_freq/sample_freq=%0.3f"%(analog_freq/sample_freq))


ax=fig.add_subplot(223)

ax.plot(x,y, color='black')
num_samples=17
sampling_x = np.linspace(0,num_samples*((y.shape[0]-1)//num_samples), num_samples).astype(np.int16) 
ax.plot(x[sampling_x] , y[sampling_x], marker='o', linewidth=0, color='red')
sample_freq=1/(x[sampling_x][1]-x[sampling_x][0])
ax.set_xlim([0,2500])
ax.set_title("analog_freq/sample_freq=%0.3f"%(analog_freq/sample_freq))

ax=fig.add_subplot(224)

ax.plot(x,y, color='black')
num_samples=10
sampling_x = np.linspace(0,num_samples*((y.shape[0]-1)//num_samples), num_samples).astype(np.int16) 
ax.plot(x[sampling_x] , y[sampling_x], marker='o', linewidth=0, color='red')
sample_freq=1/(x[sampling_x][1]-x[sampling_x][0])
ax.set_xlim([0,2500])
ax.set_title("analog_freq/sample_freq=%0.3f"%(analog_freq/sample_freq))


plt.show()
```

As you can see, when the sampling rate decreases and approaches to the frequency of the analog signal, the sampled signal (red circles) seems to have different frequency component. The phenomenon of changing frequency during sampling is called <strong>aliasing</strong>.

<strong>Sampling theorem (Shannon sampling theorem or Nyquist sampling theorem)</strong>: A continuous signal can be properly sampled only if it does not contain frequency components above one-half of the sampling rate.
